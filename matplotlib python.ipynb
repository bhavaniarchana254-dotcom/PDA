```python
#LINE PLOT
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,30,40]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#BAR CHART
import matplotlib.pyplot as plt
categories = ['A','B','C','D']
values = [5,7,3,8]
plt.bar(categories,values)
plt.title("bar chart")
plt.xlabel("catrgories")
plt.ylabel("values")
plt.show()


```

```python
#HISTOGRAM
import matplotlib.pyplot as plt
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data,bins=4)
plt.title("Histogram")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

```python
#SCATTER PLOT
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[5,7,6,8,7]
plt.scatter(x,y)
plt.title("scatter plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#OIE CHART
import matplotlib.pyplot as plt
labels = ['A','B','C','D']
sizes = [20,30,45,25]
plt.pie(sizes,labels=labels,autopct='%1.1f%%')
plt.title("pie chart")
plt.show()
```

```python
#STACKED BAR CHART
import matplotlib.pyplot as plt
x=['A','B','C']
y1 = [3,5,2]
y2 = [4,6,3]
plt.bar(x,y1,label='Data1')
plt.bar(x,y2,bottom=y1,label='Data2')
plt.title("Stacked bar chart")
plt.legend()
plt.show()         
```

```python
#STEP PLOT
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20 ,15,25]
plt.step(x,y)
plt.title("step plot")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.show()           
```

```python
#AREA PLOT
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,25]
plt.fill_between(x,y)
plt.title("area plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#BOX PLOT
import matplotlib.pyplot as plt
data = [10,20,30,40,50,100]
plt.boxplot(data)
plt.title("box plot")
plt.show()
```

```python
#STEM PLOT
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [5,15,10,20]
plt.stem(x,y)
plt.title("stem plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
